```python
def get_id(a,b,c):
    return a+b+c

class Client:
    def __init__(self, name, movie, date):
        self.name = name
        self.movie = movie
        self.date = date

class Movie:
    def __init__(self, title, start, end,quantity, price):
        self.title = title
        self.start = start
        self.end = end
        self. quantity = quantity 
        self.price = price

class Cinema:
    def __init__(self):
        from collections import defaultdict
        self.movies ={}
        self.booked = {}
        self.earnings = defaultdict(int)
        self.history = []

    def add_movie(self, title, start, end, quantity, price):
        self.movies[title] = Movie(title,start, end, quantity, price)
        self.history.append(f'Film {title} added')
        

    def remove_movie(self, title):
        if title not in self.movies:
            print('not found')
            return
        del self.movies[title]
        for key,val in self.booked.items():
            if val.movie == title:
                self.booked.pop(key)
        self.history.append(f'Film {title} was removed')

    def update_schedule(self, title, start, end):
        if title not in self.movies:
            print('Movie not found')
            return
        self.movies[title].start = start
        self.movies[title].end = end
        self.history.append(f'Schedule of film {title} was updated')

    def book_ticket(self,client, title, date):
        if title not in self.movies:
            print('movie not found')
            return
        if self.movies[title].start <= date <=  self.movies[title].end:
            self.booked[get_id(client,title,date)] = Client(client, title,date)
            self.movies[title].quantity -=1
            self.earnings[title] += self.movies[title].price
            self.history.append(f'Ticket for {client } to film {title} to date {date} was booked')
        else:
            print('unavailable')

    def refund_ticket(self, name):
        for key,val in self.booked.items():
            if val.name == name:
                title = val.movie
                self.booked.pop(key)
                self.movies[title].quantity +=1
                self.earnings[title] -= self.movies[title].price
                self.history.append(f'Ticket of {name} was refunded')
                return
            
    def total(self):
        for key, val in self.earnings.items():
            print(key,val)

    def get_history(self):
        for i in self.history:
            print(i)


bashat = Cinema()
bashat.add_movie('marvel', '10-00', '20-00', 100, 300)
bashat.book_ticket('akbar', 'marvel', '19-00')
bashat.refund_ticket('akbar')
print()
bashat.get_history()


```

```python
def get_id(a, b, c):
    return a + b + c


class Flight:
    def __init__(self, date, origin, destination, price, seats):
        self.date = date
        self.origin = origin
        self.destination = destination
        self.price = price
        self.seats = seats  # количество мест

    def change(self, new_date, new_origin, new_destination):
        self.date = new_date
        self.origin = new_origin
        self.destination = new_destination


class Passenger:
    def __init__(self, name, flight_id):
        self.name = name
        self.flight = flight_id


class Aviasales:
    def __init__(self):
        from collections import defaultdict
        self.flights = {}
        self.booked_tickets = defaultdict(list)
        self.history = []
        self.total = 0

    def add_flight(self, date, origin, destination, price, seats):
        flight_id = get_id(date, origin, destination)
        self.flights[flight_id] = Flight(date, origin, destination, price, seats)
        self.history.append(f'Flight {date} from {origin} to {destination} was added with {seats} seats')

    def remove_flight(self, date, origin, destination):
        flight_id = get_id(date, origin, destination)
        if flight_id in self.flights:
            del self.flights[flight_id]
            self.history.append(f'Flight {date} from {origin} to {destination} was removed')

    def book_flight(self, passenger, date, origin, destination):
        flight_id = get_id(date, origin, destination)
        flight = self.flights.get(flight_id)

        if flight:
            if flight.seats > 0:
                self.booked_tickets[flight_id].append(Passenger(passenger, flight_id))
                flight.seats -= 1
                self.total += flight.price
                self.history.append(f'Flight {date} from {origin} to {destination} was booked by {passenger}')
            else:
                print(f'No seats available for flight {date} from {origin} to {destination}')
        else:
            print('Flight not found')

    def cancel_booking(self, passenger, date, origin, destination):
        flight_id = get_id(date, origin, destination)
        passengers = self.booked_tickets.get(flight_id, [])
        flight = self.flights.get(flight_id)

        for p in passengers:
            if p.name == passenger:
                passengers.remove(p)
                flight.seats += 1
                self.total -= flight.price
                self.history.append(f'Booking for {passenger} on flight {flight_id} was canceled')
                return
        print('Booking not found')

    def change_flight(self, date, origin, destination, new_date, new_origin, new_destination):
        flight_id = get_id(date, origin, destination)
        if flight_id in self.flights:
            new_flight_id = get_id(new_date, new_origin, new_destination)
            flight = self.flights.pop(flight_id)
            flight.change(new_date, new_origin, new_destination)
            self.flights[new_flight_id] = flight
            self.history.append(f'Flight {date} from {origin} to {destination} was changed to {new_date} from {new_origin} to {new_destination}')

    def get_flight_schedule(self, date_range):
        for flight in self.flights.values():
            if date_range[0] <= flight.date <= date_range[1]:
                print(f'{flight.date} - {flight.origin} to {flight.destination}, {flight.seats} seats available')

    def get_available_seats(self, flight_id):
        flight = self.flights.get(flight_id)
        if flight:
            print(f'{flight.date}, {flight.origin} to {flight.destination}, Price: {flight.price}, Seats left: {flight.seats}')
        else:
            print('Flight not found')


avia = Aviasales()

avia.add_flight('2024-12-01 12:30', 'Москва', 'Нью-Йорк', 500, 100)
avia.add_flight('2024-12-05 08:00', 'Лондон', 'Токио', 700, 50)

avia.book_flight('Иван Иванов', '2024-12-01 12:30', 'Москва', 'Нью-Йорк')
avia.book_flight('Ольга Петрова', '2024-12-01 12:30', 'Москва', 'Нью-Йорк')

avia.get_available_seats(get_id('2024-12-01 12:30', 'Москва', 'Нью-Йорк'))

avia.cancel_booking('Иван Иванов', '2024-12-01 12:30', 'Москва', 'Нью-Йорк')

avia.get_available_seats(get_id('2024-12-01 12:30', 'Москва', 'Нью-Йорк'))

print(f"\nОбщая сумма бронирований: {avia.total}")
```

```python
from collections import defaultdict

class Student:
    def __init__(self, name):
        self.name = name
        self.grades = defaultdict(list)
        self.enrolled_classes = set() 


class Subject:
    def __init__(self, name):
        self.name = name
        self.grades = []


class Ebilim:
    def __init__(self):
        self.students = {}
        self.subjects = {}
        self.history = []

    def add_student(self, name):
        if name in self.students:
            print(f"Student {name} already exists.")
            return
        self.students[name] = Student(name)
        self.history.append(f"Student {name} was added.")

    def remove_student(self, student):
        if student in self.students:
            del self.students[student]
            self.history.append(f"Student {student} was removed.")
        else:
            print(f"Student {student} not found.")

    def add_grade(self, student, sabak, grade, date):
        if student not in self.students:
            print(f"Student {student} not found.")
            return
        if sabak not in self.subjects:
            print(f"Subject {sabak} not found.")
            return
        self.students[student].grades[sabak].append((date, grade))
        self.history.append(f"Grade {grade} for {sabak} was added to {student} on {date}.")

    def update_grade(self, student, sabak, grade, date):
        if student in self.students and sabak in self.students[student].grades:
            for idx, entry in enumerate(self.students[student].grades[sabak]):
                if entry[0] == date:
                    self.students[student].grades[sabak][idx] = (date, grade)
                    self.history.append(f"Grade for {student} in {sabak} on {date} was updated to {grade}.")
                    return
        print('Grade not found.')

    def add_class(self, sabak):
        if sabak in self.subjects:
            print(f"Subject {sabak} already exists.")
            return
        self.subjects[sabak] = Subject(sabak)
        self.history.append(f"Class {sabak} was added.")

    def remove_class(self, sabak):
        if sabak in self.subjects:
            del self.subjects[sabak]
            self.history.append(f"Class {sabak} was removed.")
        else:
            print(f"Class {sabak} not found.")

    def register_to_class(self, student, sabak):
        if student not in self.students:
            print(f"Student {student} not found.")
            return
        if sabak not in self.subjects:
            print(f"Subject {sabak} not found.")
            return
        self.students[student].enrolled_classes.add(sabak)
        self.history.append(f"Student {student} registered to {sabak}.")

    def unregister_from_class(self, student, sabak):
        if student not in self.students:
            print(f"Student {student} not found.")
            return
        if sabak not in self.students[student].enrolled_classes:
            print(f"Student {student} is not registered to {sabak}.")
            return
        self.students[student].enrolled_classes.remove(sabak)
        self.history.append(f"Student {student} unregistered from {sabak}.")

    def get_history(self):
        for event in self.history:
            print(event)
```
