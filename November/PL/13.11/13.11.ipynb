```python
from PyQt5.QtWidgets import QApplication, QMainWindow, QLabel, QLineEdit, QPushButton, QTextEdit

class Client:
    def __init__(self, name):
        self.name = name
        self.total = 0

    def getBalance(self):
        return self.total

    def add(self, amount):
        self.total += amount

    def remove(self, amount):
        self.total -= amount


class Optima(QMainWindow):
    def __init__(self):
        super().__init__()

        self.clients = {}
        self.total = 0
        self.history = []

        self.initUI()

    def initUI(self):
        self.setWindowTitle('Optima Bank')

        # Create labels and input fields
        nameLabel = QLabel('Name:', self)
        nameLabel.move(10, 10)
        self.nameInput = QLineEdit(self)
        self.nameInput.move(80, 10)
        self.nameInput.resize(200, 30)

        phoneLabel = QLabel('Phone:', self)
        phoneLabel.move(10, 50)
        self.phoneInput = QLineEdit(self)
        self.phoneInput.move(80, 50)
        self.phoneInput.resize(200, 30)

        amountLabel = QLabel('Amount:', self)
        amountLabel.move(10, 90)
        self.amountInput = QLineEdit(self)
        self.amountInput.move(80, 90)
        self.amountInput.resize(200, 30)

        # Create buttons
        addButton = QPushButton('Add', self)
        addButton.move(10, 130)
        addButton.clicked.connect(self.addClient)

        transferButton = QPushButton('Transfer', self)
        transferButton.move(10, 170)
        transferButton.clicked.connect(self.transfer)

        balanceButton = QPushButton('Balance', self)
        balanceButton.move(10, 210)
        balanceButton.clicked.connect(self.getBalance)

        # Create text area for history
        self.historyText = QTextEdit(self)
        self.historyText.move(10, 250)
        self.historyText.resize(300, 200)

        self.show()

    def addClient(self):
        name = self.nameInput.text()
        phone = self.phoneInput.text()

        self.clients[name] = Client(name)
        self.history.append(f'Client {name} add with phone number {phone}')

        self.updateHistory()

    def transfer(self):
        to_name = self.nameInput.text()
        from_name = self.phoneInput.text()
        amount = float(self.amountInput.text())

        komissiya = amount * 0.1/100
        self.clients[to_name].add(amount)
        self.clients[from_name].remove(amount+komissiya)
        self.total += komissiya
        self.history.append(f'Transfer {amount} from {from_name} to {to_name}')

        self.updateHistory()

    def getBalance(self):
        name = self.nameInput.text()
        balance = self.clients[name].getBalance()
        self.history.append(f'Balance of {name}: {balance}')

        self.updateHistory()

    def updateHistory(self):
        self.historyText.setText('\n'.join(self.history))


if __name__ == '__main__':
    app = QApplication([])
    window = Optima()
    app.exec_()
```
